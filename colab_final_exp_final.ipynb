```python
import re
def clean_sampled(text):

    if isinstance(text,str):
        text = text.lower()

    pattern1 = re.compile(r'\n')
    text = re.sub(pattern1,' ',text)

    pattern2 = re.compile(r'[!@!#$%^&*()_+|]')
    text = re.sub(pattern2,'',text)

    text = text.strip()




    return text




```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, Trainer, TrainingArguments
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training
from datasets import Dataset
import torch
import pandas as pd
from sklearn.model_selection import train_test_split
from transformers import BitsAndBytesConfig

class ChatbotModelTrainer:
    def __init__(self, model_path='zementalist/llama-3-8B-chat-psychotherapist', max_length=512, quantize=True):
        # Load the tokenizer
        self.tokenizer = AutoTokenizer.from_pretrained(model_path)
        self.tokenizer.pad_token = self.tokenizer.eos_token
        self.max_length = max_length

        # Enable 8-bit quantization using BitsAndBytes
        if quantize:
            bnb_config = BitsAndBytesConfig(load_in_8bit=True)  # Enable 8-bit quantization
            # Load the model with quantization
            self.model = AutoModelForCausalLM.from_pretrained(
                model_path,
                quantization_config=bnb_config,
                device_map="auto",  # Automatic device placement
                low_cpu_mem_usage=True
            )
        else:
            self.model = AutoModelForCausalLM.from_pretrained(
                model_path,
                device_map="auto",
                low_cpu_mem_usage=True
            )

        # Freeze most layers
        for name, param in self.model.named_parameters():
            if "lora" not in name and "embeddings" not in name and "attn" not in name:
                param.requires_grad = False

        # LoRA configuration
        peft_config = LoraConfig(
            lora_alpha=16,
            lora_dropout=0.1,
            r=8,
            bias="none",
            task_type="CAUSAL_LM",
            target_modules=["q_proj", "v_proj"]
        )

        # Prepare model for k-bit training
        self.model = prepare_model_for_kbit_training(self.model)
        self.model = get_peft_model(self.model, peft_config)

        # Print trainable parameters
        self.model.print_trainable_parameters()

    def prepare_dataset(self, dataframe):
        dataframe['input_text'] = (
            "### Human: " + dataframe['Question'] + "\n### Assistant: " + dataframe['Answer']
        )

        train_texts, val_texts = train_test_split(dataframe['input_text'], test_size=0.2, random_state=42)

        train_encodings = self.tokenizer(train_texts.tolist(), truncation=True, padding=True, max_length=self.max_length, return_tensors='pt')
        val_encodings = self.tokenizer(val_texts.tolist(), truncation=True, padding=True, max_length=self.max_length, return_tensors='pt')

        train_labels = train_encodings['input_ids'].clone()
        val_labels = val_encodings['input_ids'].clone()

        train_dataset = Dataset.from_dict({
            'input_ids': train_encodings['input_ids'],
            'attention_mask': train_encodings['attention_mask'],
            'labels': train_labels
        })

        val_dataset = Dataset.from_dict({
            'input_ids': val_encodings['input_ids'],
            'attention_mask': val_encodings['attention_mask'],
            'labels': val_labels
        })

        return train_dataset, val_dataset

    def train(self, train_dataset, val_dataset):
        training_args = TrainingArguments(
            output_dir='./therapy_model_results',
            num_train_epochs=3,
            per_device_train_batch_size=5,  # Reduced batch size
            per_device_eval_batch_size=5,   # Reduced batch size
            gradient_accumulation_steps=8,  # Accumulate gradients
            warmup_steps=50,
            learning_rate=5e-4,
            weight_decay=0.01,
            logging_dir='./logs',
            logging_steps=10,
            eval_strategy='epoch',
            fp16=True,  # Mixed precision
            report_to='none',
            max_grad_norm=0.3,
            save_total_limit=2,
            dataloader_num_workers=4,  # Increase num_workers to load data faster
           no_cuda=False,
        )

        trainer = Trainer(
            model=self.model,
            args=training_args,
            train_dataset=train_dataset,
            eval_dataset=val_dataset
        )

        trainer.train()
        return trainer

    def save_model(self, trainer):
        trainer.save_model()
        self.tokenizer.save_pretrained(trainer.state.output_dir)

```

```python
from transformers import AutoModelForCausalLM, Trainer

def save_model_and_tokenizer_from_trainer(trainer, output_dir="./saved_model"):
    """
    Save the model and tokenizer from the Trainer instance to the specified directory.

    Args:
    - trainer: The Trainer object containing the model and tokenizer.
    - output_dir: The directory to save the model and tokenizer.
    """
    # Save the model
    trainer.save_model(output_dir)

    # Access tokenizer via processing_class
    tokenizer = trainer.processing_class.tokenizer
    if tokenizer:
        tokenizer.save_pretrained(output_dir)
        print(f"Tokenizer saved to {output_dir}")
    else:
        print("Tokenizer not found in Trainer instance. Please provide a valid tokenizer.")

    print(f"Model saved to {output_dir}")

```

```python
import gradio as gr
from transformers import AutoTokenizer, AutoModelForCausalLM, AutoModelForSequenceClassification, pipeline
import torch
from transformers import BitsAndBytesConfig  # For quantization

# Paths to the saved models and tokenizer
chatbot_model_path = "brightdatum/bright_therapy_model_3.0_8B"  # Therapy chatbot model
classifier_model_path = "brightdatum/suicideclassiferbert1.0"  # Suicide classification model

# Load the tokenizer for the chatbot
chatbot_tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-3.1-8B")

# Ensure tokenizer has a pad token
chatbot_tokenizer.pad_token = chatbot_tokenizer.eos_token

# Configure and load the quantized chatbot model
quantization_config = BitsAndBytesConfig(llm_int8_enable_fp32_cpu_offload=True)  # Using int8 quantization
chatbot_model = AutoModelForCausalLM.from_pretrained(
    chatbot_model_path,
    device_map="auto",  # Auto-detect device
    quantization_config=quantization_config,
    torch_dtype=torch.float16  # Use mixed precision
)

# Set the chatbot model to evaluation mode
chatbot_model.eval()

# Load the suicide classification model
classifier = pipeline(
    "text-classification",
    model=classifier_model_path,
    tokenizer=classifier_model_path,
    device=0  # Assuming GPU; change to -1 for CPU
)

def classify_and_respond(user_input):
    """
    First classify the input for suicidal intent, then respond accordingly.
    """
    # Classify the input
    classification = classifier(user_input)
    label = classification[0]['label']
    score = classification[0]['score']

    # Default response for suicidal intent
    if label == "suicidal" and score > 0.8:  # Threshold for confident classification
        return (
            """ I cannot provide you with assistance in ending your life. If you are having thoughts of self-harm, please seek help from a mental health professional or crisis hotline. There are many resources available to help you, including:
                1. National Suicide Prevention Lifeline (in the United States): 1-800-273-TALK (8255)
                2. Crisis Text Line (in the United States): Text HOME to 741741"""
        )

    # If not suicidal, proceed with chatbot response
    input_text = f"### Human: {user_input}\n### Assistant:"
    inputs = chatbot_tokenizer(input_text, return_tensors="pt", truncation=True, padding=True, max_length=512).to("cuda")

    # Generate response
    with torch.no_grad():
        outputs = chatbot_model.generate(
            inputs["input_ids"],
            max_length=512,
            num_return_sequences=1,
            no_repeat_ngram_size=2,  # Avoid repetitive phrases
            top_p=0.9,              # Restrict to high-probability tokens
            temperature=0.8,        # Add some randomness (lower for deterministic)
            pad_token_id=chatbot_tokenizer.pad_token_id,
        )

    # Decode and clean the response
    decoded_output = chatbot_tokenizer.decode(outputs[0], skip_special_tokens=True)
    response = decoded_output.split("### Assistant:")[-1].strip()
    return response

# Define custom styling
css = """
body {
    background-color: #f7f7f7; /* Light gray background */
    font-family: Arial, sans-serif;
}

.gr-button {
    background-color: #0078D4; /* Light blue button */
    color: white;
    border-radius: 5px;
}

.gr-button:hover {
    background-color: #005A9E; /* Darker blue on hover */
}
"""

# Create a Gradio interface with simple styling
iface = gr.Interface(
    fn=classify_and_respond,
    inputs="text",
    outputs="text",
    title="Therapy Chatbot",
    description="A simple chatbot for therapy-like responses. Type your thoughts below.",
    theme="compact",  # Compact theme
    css=css,  # Add custom styling
)

# Launch the interface
iface.launch()
```

```python
import pandas as pd
import numpy as np
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
from nltk.translate.bleu_score import corpus_bleu
from sklearn.metrics.pairwise import cosine_similarity
from collections import Counter

# Load model and tokenizer once
model_name = "brightdatum/therapychatbot2"
model = AutoModelForCausalLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token
model.config.pad_token_id = tokenizer.pad_token_id

def evaluate_model(dataset, sample_size=50):
    """
    Evaluate model and produce structured JSON results.
    
    Args:
        dataset (pd.DataFrame): Dataset with 'Question' and 'Answer' columns
        sample_size (int): Number of samples for evaluation

    Returns:
        Dict with evaluation metrics and saves results as JSON file.
    """
    sampled_data = dataset.sample(min(sample_size, len(dataset)))
    references = [[answer.split()] for answer in sampled_data['Answer']]
    inputs = [
        tokenizer(f"Question: {row['Question']}\nAnswer:", return_tensors="pt", max_length=512, truncation=True)
        for _, row in sampled_data.iterrows()
    ]

    predictions, perplexities, avg_lengths, unique_ratios, similarity_scores = [], [], [], [], []

    for input_data, (_, row) in zip(inputs, sampled_data.iterrows()):
        with torch.no_grad():
            output = model.generate(**input_data, max_length=512, num_return_sequences=1, do_sample=False)
            full_output = tokenizer.decode(output[0], skip_special_tokens=True)
        
        answer_start = full_output.find("Answer:") + len("Answer:")
        prediction = full_output[answer_start:].strip().split()
        predictions.append(prediction)

        # Calculate perplexity
        try:
            input_ids = tokenizer(full_output, return_tensors="pt")['input_ids']
            with torch.no_grad():
                loss = model(input_ids, labels=input_ids).loss
                perplexities.append(torch.exp(loss).item())
        except Exception:
            perplexities.append(np.nan)

        # Text quality metrics
        words = prediction
        avg_lengths.append(np.mean([len(word) for word in words]) if words else 0)
        unique_ratios.append(len(set(words)) / len(words) if words else 0)

        # Cosine similarity (simple embedding approach for demonstration)
        question_vector = tokenizer(row['Question'], return_tensors="pt")['input_ids'].mean(dim=1)
        pred_vector = tokenizer(' '.join(prediction), return_tensors="pt")['input_ids'].mean(dim=1)
        similarity_scores.append(
            cosine_similarity(question_vector, pred_vector)[0][0] if len(question_vector) == len(pred_vector) else 0
        )

    bleu_score = corpus_bleu(references, predictions)

    results = {
        "generation_metrics": {
            "perplexity": np.nanmean(perplexities),
            "generation_length": np.mean([len(pred) for pred in predictions])
        },
        "text_quality_metrics": {
            "unique_words_ratio": np.mean(unique_ratios),
            "avg_word_length": np.mean(avg_lengths)
        },
        "comparison_metrics": {
            "exact_match": 0.0,  # Placeholder; compute if exact match metric is relevant
            "similarity_score": np.mean(similarity_scores)
        }
    }

    # Save results to JSON
    import json
    with open("evaluation_results.json", "w") as f:
        json.dump(results, f, indent=4)

    print("Evaluation complete. Results saved to 'evaluation_results.json'.")
    return results


results = evaluate_model(dataset, sample_size=50)
```
